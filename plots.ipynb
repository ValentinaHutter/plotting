```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

data = xr.open_dataset("WGS84_E15N47_20190702T120000.tif")

data = data.band_data

brg = np.zeros((data[0].shape[0],data[0].shape[1],3))
brg[:,:,0] = (data[2].values+1)/6.8
brg[:,:,1] = (data[1].values+1)/6.4
brg[:,:,2] = (data[0].values+1)/5.3
brg = np.clip(brg,0,255).astype(np.uint8)

plt.figure(figsize=(8,12))
plt.title("RGB Austria, July 2019")
plt.imshow(brg,cmap='brg')
```
